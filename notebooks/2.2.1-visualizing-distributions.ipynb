## Plotting distributions using a histograms

### Creating data

Let's begin by creating a random distribution with a mean of 100 ($\mu=100$) and a variance of 15 ($\sigma=15$)


```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

%matplotlib inline 

mu, sigma = 100, 15
x = mu + sigma*np.random.randn(10000)

print "Actual Mean: " + str(np.mean(x))
print "Actual Standard Deviation:" + str(np.std(x))
```

```python
plt.plot(x, 'green')
plt.xlabel("Measurement Number")
plt.ylabel("IQ")
plt.title("IQ measurements")
```

### Plotting Univeriate Data

The most basic visualization to plot univariate data is to use a histrogram

```python
# the histogram of the data
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='green', alpha=0.75, label="Actual Distribution")


plt.legend()
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('$\mathrm{Histogram\ of\ IQ:}\ \mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)

plt.show()
```

We can also plot the theoretical normal distribution with the same mean and standard deviation.

```python
# the histogram of the data
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='green', alpha=0.75, label="Actual Distribution")

# add a 'best fit' line
y = mlab.normpdf( bins, mu, sigma)
plt.plot(bins, y, 'r--', linewidth=1, label="Theoretical Distribution")

plt.legend()
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('$\mathrm{Histogram\ of\ IQ:}\ \mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
```

We can also use the seaborn plotting library  default, this will draw a histogram and fit a kernel density estimate (KDE).

```python
import seaborn as sns

sns.distplot(x);
plt.xlabel('Smarts')
plt.ylabel('Probability')
```

## Alternative plots for distributions

### Rug Plot

A tick for every measurements


```python
sns.distplot(x, kde=False, rug=True);
```

```python
sns.distplot(x, hist=False, rug=True);
```

```python
sns.distplot(x, kde=True, rug=True);
```

## QQ-Plot

```python
import scipy.stats as stats
import pylab

stats.probplot(x, dist="norm", plot=pylab)
```

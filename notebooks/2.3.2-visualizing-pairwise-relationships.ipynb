## Plotting Pairplots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


%matplotlib inline 

iris = pd.read_csv('../datasets/iris.csv')
```

```python
iris.columns
```

```python
iris.describe()
```

```python
iris.plot()
```

```python
sns.pairplot(iris, hue="iris")
```

We can combine a KD plot with a scatter plot.

```python
g = sns.PairGrid(iris,hue="iris")
g.map_lower(sns.kdeplot)
g.map_upper(plt.scatter)
g.map_diag(sns.kdeplot, lw=3)
```

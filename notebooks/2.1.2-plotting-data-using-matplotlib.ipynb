### Importing matplotlib.pyplot

`pyplot` is a collection of command style functions that makes plotting easy with matplotlib!

Here we import it as `plt`

````python
import matplotlib.pyplot as plt
````

Matplotlib works seamlessly with Jupyter notebook, but it must be setup first using the following magic function:

````python
%matplotlib notebook
````

`%matplotlib notebook` enables interactive plots, while `%matplotlib inline` enables static plots (images)



````python
%matplotlib inline
````

## Plotting data traces using the plot function


We can now plot the columns of the dataframe

````python
plt.plot(anscombe_i["x"])
````

````python
plt.plot(anscombe_i["y"])
````

````python
plt.plot(anscombe_i["x"], label="x")
plt.plot(anscombe_i["y"], label="y")
plt.legend()
````

````python
anscombe_i.plot()
````

## Labeling matplotlib figures

It is not enough to just plot data, one must label the figure.  Without the axis labeled, no figure is complete.


````python
anscombe_i.plot()
plt.xlabel("Measurement Number")
plt.ylabel("Measurement Value (unitless)")
plt.title("Trace of X and Y for Anscombe's first dataset")
````

## Plotting multivariate data (scatter plot)

Now let's plot the x and y against each other.

````python
plt.plot(anscombe_i["x"],anscombe_i["y"],'.')

# Equivilent to above
#plt.scatter(anscombe_i["x"],anscombe_i["y"])

plt.xlabel("Measurement of X (unitless)")
plt.ylabel("Measurement of Y (unitless)")
plt.title("Y vs X for Anscombe's first dataset")
````

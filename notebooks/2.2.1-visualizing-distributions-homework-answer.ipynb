Plot the distribution of the readings in `uniform_random_numbers.csv`.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import pylab

%matplotlib inline
```

```python
uniform_random_numbers_df = pd.read_csv("../datasets/uniform_random_numbers.csv")
uniform_random_numbers = uniform_random_numbers_df["random_number"]
plt.plot(uniform_random_numbers)
plt.xlabel("Measurement Number")
plt.ylabel("Random Number")
```

```python
uniform_random_numbers.describe()
```

```python
sns.distplot(uniform_random_numbers, kde=False);
plt.xlabel("Random Number")
plt.ylabel("Frequency")
```

```python
stats.probplot(uniform_random_numbers, dist="norm", plot=pylab)
```

## Plotting Anscobe's first dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


%matplotlib inline 

anscombe_i = pd.read_csv('../datasets/anscombe_i.csv')
```

```python
plt.plot(anscombe_i)
```

```python
anscombe_i.describe()
```

The simpliest way to visualize a bivariate distribution is a scatterplot, where each observation is shown with point at the x and y values. This is analgous to a rug plot (from plotting univariate distributions) on two dimensions. 

```python
plt.scatter(anscombe_i["x"],anscombe_i["y"])
plt.xlabel("x")
plt.ylabel("y")
```

You can draw a scatterplot with the matplotlib plt.scatter function, and it is also the default kind of plot shown by the jointplot() function along with a histrogram for each variable.

```python
sns.jointplot(x="x", y="y", data=anscombe_i)
```

Bivariate distributions can also be done with a kernel density estimation using a contour plot:

```python
sns.jointplot(x="x", y="y", data=anscombe_i, kind="kde");
```

We can also pair the kernal density plot with a rug plot for each of the two variables.

```python
f, ax = plt.subplots(figsize=(6, 6))
sns.kdeplot(anscombe_i.x, anscombe_i.y, ax=ax)
sns.rugplot(anscombe_i.x, color="g", ax=ax)
sns.rugplot(anscombe_i.y, vertical=True, ax=ax);
```

```python
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline

anscombe_i = pd.read_csv('../datasets/anscombe_i.csv')

fig = anscombe_i.plot()
plt.xlabel("Measurement Number")
plt.ylabel("Measurement Value (unitless)")
plt.title("Trace of X and Y for Anscombe's first dataset")
```



## Choosing a style

After importing matplotlib, we can view the avaliable styles.


```python
print(plt.style.available)
```

```python
# change figure size
fig = anscombe_i.plot(figsize=(4,3))
plt.xlabel("Measurement Number")
plt.ylabel("Measurement Value (unitless)")
plt.title("Trace of X and Y for Anscombe's first dataset")
```

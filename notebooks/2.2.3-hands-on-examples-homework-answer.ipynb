## Josephson Junction Cryothermometry (homework)

Taken from: http://www.itl.nist.gov/div898/handbook/eda/section4/eda424.htm

This data set was collected by Bob Soulen of NIST in October, 1971 as a sequence of observations collected equi-spaced in time from a volt meter to ascertain the process temperature in a Josephson junction cryothermometry (low temperature) experiment. The response variable is voltage counts.

Conduct exploratory data analysis on the univariate dataset.


## Answer

Must haves:

1. Legend
2. Labeled Axis and Title
3. Labels on datasets
4. Visually distinguishable


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import pylab


%matplotlib inline

cryothermometry_df = pd.read_csv("../datasets/joseph-junction-cryothermometry.csv")
cryothermometry = cryothermometry_df["voltage_counts"]

cryothermometry.plot()
plt.xlabel("Measurement Number")
plt.ylabel("voltage counts")
```

```python
sns.distplot(cryothermometry, kde=False);
plt.xlabel("Filter Transmittance")
plt.ylabel("Frequency")
```

```python
sns.boxplot(cryothermometry)
```

```python
stats.probplot(cryothermometry, dist="norm", plot=pylab)
```
